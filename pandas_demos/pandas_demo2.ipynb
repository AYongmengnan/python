```python
# 设置为inline风格
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tuples = list(zip(*[['bar','bar','baz','baz',
                     'foo','foo','qux','qux'],
                   ['one','two','one','two',
                    'one','two','one','two']]))
```

```python
index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
index
```

```python
df = pd.DataFrame(np.random.randn(8,2), index=index, columns=['A','B'])
df
```

```python
stacked = df.stack()
stacked
```

```python
stacked = df.unstack()
stacked
```

```python
stacked = df.unstack().unstack()
stacked
```

```python
#数据透视表,值是A这一列，行索引为first，列索引为second
df.pivot_table(values='A', index=['first'], columns=['second'])
```
